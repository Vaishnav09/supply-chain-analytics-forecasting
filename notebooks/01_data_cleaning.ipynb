```python
df["order_date"]    = pd.to_datetime(df["order date (DateOrders)"], errors="coerce")
df["shipping_date"] = pd.to_datetime(df["shipping date (DateOrders)"], errors="coerce")
```

```python
print(df[['order_date', 'shipping_date']].dtypes)
print(df.isnull().sum().sum(), "missing values remaining")
```

```python
#Clean column names
df = df.drop(columns=[
    "Product Description",
    "Order Zipcode",
    "Customer Email",
    "Customer Password",
    "Customer Fname",
    "Customer Lname",
    "Customer Street",
    "Product Image"
])
```

```python
df.isnull().sum().sum()
```

```python
df.shape
```

```python
df.columns
```

```python
#clean column names
df.columns = (df.columns.str.strip()
            .str.lower()
            .str.replace(' ', '_')
            .str.replace('(', '')
            .str.replace(')', '')
            )
```

```python
df.dtypes
```

```python
# Create shipping delay
df['shipping_delay'] = (df['days_for_shipping_real'] - df['days_for_shipment_scheduled'])
```

```python
df['shipping_delay'].describe()
```

```python
# Convert numeric columns
df["order_item_total"] = pd.to_numeric(
    df["order_item_total"], errors="coerce"
    )
```

```python
# order_month and is_late are derived columns
df["order_month"] = df["order_date"].dt.to_period("M").astype(str)
df["is_late"]     = df["late_delivery_risk"] == 1
```

```python
df.to_csv("../data/processed/clean_supply_chain.csv", index=False)
print(f"saved {len(df):,} rows x {len(df.columns)} columns")
print("order_date range:", df["order_date"].min().date(), "to", df["order_date"].max().date())
```
